```python
import sqlite3
```

```python
conn = sqlite3.connect("INSTRUCTOR.db")
```

```python
cursor_obj = conn.cursor()
```

```python
cursor_obj.execute("DROP TABLE IF EXISTS INSTRUCTOR")
```

```python
table = """ create table IF NOT EXISTS INSTRUCTOR(ID INTEGER PRIMARY KEY NOT NULL, FNAME VARCHAR(20), LNAME VARCHAR(20), CITY VARCHAR(20), COODE CHAR(2));"""

cursor_obj.execute(table)

print("Table is ready")
```

```python
cursor_obj.execute("""insert into INSTRUCTOR values (2, "Raul", "Chong", "Markham", "CA"), (3, "Hima", "Vasudevan", "Chicago", "US");""")
```

```python
statement = """SELECT * FROM INSTRUCTOR"""
cursor_obj.execute(statement)

print("All the data")
output_all = cursor_obj.fetchall()
for row_all in output_all:
    print(row_all)
```

```python
statement = """SELECT FNAME FROM INSTRUCTOR"""
cursor_obj.execute(statement)

print("All the data")
output1 = cursor_obj.fetchmany(2)
for row in output1:
    print(row)
```

```python
statement = """SELECT CITY FROM INSTRUCTOR"""
cursor_obj.execute(statement)

print("All the data")
output2 = cursor_obj.fetchmany(2)
for out in output2:
    print(out)
```
