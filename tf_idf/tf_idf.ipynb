```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = [
    "Common cold symptoms include a runny nose, sneezing, and a sore throat.",
    "Influenza, or the flu, often causes high fever, muscle aches, and fatigue.",
    "Diabetes is a chronic condition that affects how the body processes blood sugar.",
    "Asthma is a respiratory condition marked by spasms in the bronchi of the lungs.",
    "Hypertension is another name for high blood pressure which can lead to heart disease."
]
```

```python
df = pd.DataFrame(data)
df
```

```python
tfidf_model = TfidfVectorizer()
```

```python
data_vectors = tfidf_model.fit_transform(data)
```

```python
model_set = tfidf_model.get_feature_names_out()
```

```python
vector_rep = data_vectors.toarray()
```

```python
vector_rep
```

```python
df_tf_idf = pd.DataFrame(vector_rep, columns = model_set)
df_tf_idf
```

```python
ave_tf_idf = df_tf_idf.mean(axis=0)
ave_tf_idf
```
