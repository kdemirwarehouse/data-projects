```python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse_output=False)
encoded_array = one_hot_encoder.fit_transform(df[["gender"]])
encoded_array
```

```python
#let's get new column names
encoded_columns = one_hot_encoder.get_feature_names_out(["gender"])
```

```python
#let's create new dataframe
encoded_df = pd.DataFrame(encoded_array, columns=encoded_columns)
encoded_df
```

```python
#let's merge dataframes
df_final = pd.concat([df, encoded_df], axis=1)
df_final
```

```python
#dummy variable trick
encoded_df = encoded_df.drop(columns=["gender_Female"])
encoded_df
```
